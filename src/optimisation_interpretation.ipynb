# The plot of the paper was Coverage vs Response Time 

```python
sns.lineplot(df_report[['r2','alpha']],x='r2',y='alpha',errorbar=None,color='blue',label='r2')
sns.lineplot(df_report[['r1','alpha']],x='r1',y='alpha',errorbar=None,color='red',label='r1')
plt.title("Coverage () vs Secondary Response Time (r2)")
plt.xlabel('r1/r2 Response Time')
plt.ylabel('Coverage ()')
plt.xticks(np.arange(5,30,1))
plt.yticks(np.arange(0.8,1.01,0.1))
plt.grid(True)
plt.show()
```
