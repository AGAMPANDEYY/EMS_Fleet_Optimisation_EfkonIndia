```python
# Function to find nearest value
def find_nearest_value(x, df):
    list_diff = []
    for row in df:
        diff_val = abs(x - row)
        list_diff.append(diff_val)
    nearest_value = df[list_diff.index(min(list_diff))]
    return nearest_value
```

```python
df2['CHAINAGE']
```

```python
# Iterate through rows in sorted_df to find nearest equipment chainage
nearest_values = []
for index, row in sorted_df.iterrows():
    val_col1 = row['PPT_Chainage No']
    nearest_val_col2 = find_nearest_value(val_col1, df2['CHAINAGE'])
    nearest_values.append(nearest_val_col2)

# Add nearest equipment chainage to sorted_df
sorted_df['nearest_equip_chainage'] = nearest_values

# Display the final dataframe
df=sorted_df[['PPT_Chainage No', 'nearest_equip_chainage']]
```

## Calculating the max and min of the distance between each equip chainage

```python
max_distance=0
min_distance=0
for i in range(0,len(df2)-1):
    equip1_chainage=df2.loc[i,'CHAINAGE']
    equip2_chainage=df2.loc[i+1,'CHAINAGE']
    print(equip1_chainage)
    print(equip2_chainage)
    distance=abs(equip2_chainage-equip1_chainage)
    if distance!=0:
        if distance>max_distance:
            max_distance=distance
        elif distance<min_distance:
            min_distance=distance
    print(f"Min Distance {min_distance}")
    print(f"Max Distance {max_distance}")
    print("--------------------")
```

## Calculating Accident Actual Lat/Long

```python
degree_to_km = 111.111
df=pd.read_csv("/home/mint/Desktop/Agam_Strabag/files/equipment_lat_lon_fatal.csv")
accident_lat_list=[]
accident_lon_list=[]


for i, row in df.iterrows():

    equip_lat=row['Equip_LATITUDE']
    equip_lon=row['Equip_LONGITUDE']
    diff_acc_equipkm=row['Accident_Chainage']-row['Equipment_Chainage']
    if diff_acc_equipkm==0:
          accident_lat_list.append(equip_lat)
          accident_lon_list.append(equip_lon)
          continue
    else:
      change_in_lat=(diff_acc_equipkm/111.11)
      change_in_lon=(diff_acc_equipkm/111.11)

      accident_lat=row['Equip_LATITUDE']+change_in_lat
      accident_lon=row['Equip_LONGITUDE']+change_in_lon
      accident_lat_list.append(accident_lat)
      accident_lon_list.append(accident_lon)

    print(f"Equiment Chainage {row['Equipment_Chainage']}| Accident Chainage {row['Accident_Chainage']}")
    print(f"Euipment lat is {equip_lat} | Accident lat is {accident_lat}")
    print(f"Euipment lon is {equip_lon} | Accident lon is {accident_lon}")
    print("---------------------------------------------------------------------")
```

```python
df['Accident_LATITUDE']=accident_lat_list
df['Accident_LONGITUDE']=accident_lon_list 
```
